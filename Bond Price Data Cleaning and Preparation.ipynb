### shift EVAL MID PRICE one month earlier as the output

```python
def series_to_supervise(df, s=-1):
    shifted_df = df.shift(s)
    shifted_df.rename(columns= {'EVAL MID PRICE':'OUTPUT PRICE'}, inplace=True)
    y_out = shifted_df['OUTPUT PRICE']
    del shifted_df
    
    return pd.concat([df, y_out], axis=1)
```

```python
grouped = df.groupby(['STOCK CODE'])
grouped
```

```python
final_data = []
for name, data in grouped:
    data_sup = series_to_supervise(data)
    data_sup.dropna(subset=['OUTPUT PRICE'], inplace=True)
    final_data.append(data_sup)
```

```python
df_sup = pd.concat(final_data)
```

```python
def convertMonths(r):
    if r[-1].__eq__('+'):
        r = r[:-1]
    
    if r[-1].__eq__('Y'):
        return int(r[:-1])*12
    elif r[-1].__eq__('M'):
        return int(r[:-1])
              
```

```python
df['REMAINING TENURE'] = df['REMAINING TENURE'].apply(lambda x : convertMonths(x))
```
